```python
##Editor: Qiuming Li
import os
import cv2
import numpy as np
import random

def sp_noise(noise_img, proportion):
    
    height, width = noise_img.shape[0], noise_img.shape[1]#get the pixel value
    num = int(height * width * proportion) #to put the number of noise points
    for i in range(num):
        w = random.randint(0, width - 1)
        h = random.randint(0, height - 1)
        if random.randint(0, 1) == 0:
            noise_img[h, w] = 0
        else:
            noise_img[h, w] = 255
    return noise_img

def convert(input_dir, output_dir):
    for filename in os.listdir(input_dir):
        path = input_dir + "/" + filename # get the file path
        print("doing... ", path)
        noise_img = cv2.imread(path)
        img_noise = sp_noise(noise_img,0.025)# sp noise
        cv2.imwrite(output_dir+'/'+filename,img_noise )

if __name__ == '__main__':
    input_dir = "C:/Users/49162/Desktop/python/images/camera"    # 输入数据文件夹
    output_dir = "C:/Users/49162/Desktop/python/images/sp_noise" # 输出数据文件夹
    convert(input_dir, output_dir)
```

```python
import os
import cv2
import numpy as np
import random

def gaussian_noise(img, mean, sigma):
   
    # normalization the pictures
    img = img / 255
    # make gaussian noise
    noise = np.random.normal(mean, sigma, img.shape)
    gaussian_out = img + noise
    gaussian_out = np.clip(gaussian_out, 0, 1)
    gaussian_out = np.uint8(gaussian_out*255)
    return gaussian_out

def convert(input_dir, output_dir):
    for filename in os.listdir(input_dir):
        path = input_dir + "/" + filename 
        print("doing... ", path)
        noise_img = cv2.imread(path)
        img_noise = gaussian_noise(noise_img, 0, 0.44) 
        cv2.imwrite(output_dir+'/'+filename,img_noise )

if __name__ == '__main__':
    input_dir = "C:/Users/49162/Desktop/python/images/depth_real"    
    output_dir = "C:/Users/49162/Desktop/python/images/gaussian_noise_depth_sigma40" 
    convert(input_dir, output_dir)
```

```python
import os
import cv2
import numpy as np
import random

def random_noise(image,noise_num):
   
    img_noise = image
    rows, cols, chn = img_noise.shape
    for i in range(noise_num):
        x = np.random.randint(0, rows)
        y = np.random.randint(0, cols)
        img_noise[x, y, :] = 255
    return img_noise


def convert(input_dir, output_dir):
    for filename in os.listdir(input_dir):
        path = input_dir + "/" + filename 
        print("doing... ", path)
        noise_img = cv2.imread(path)
        img_noise  = random_noise(noise_img,500)
        cv2.imwrite(output_dir+'/'+filename,img_noise )

if __name__ == '__main__':
    input_dir = "C:/Users/49162/Desktop/python/images/camera"    
    output_dir = "C:/Users/49162/Desktop/python/images/random_noise" 
    convert(input_dir, output_dir)
```
